### The + operator does not support set addition (union) directly in Python. Instead, you can use set operations such as union() or the | operator to combine sets. Here's how you can work with sets and other data types in Python:

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

# Using union() method
result = set1.union(set2)
print("Union of sets using union():", result)  # Output: {1, 2, 3, 4, 5}

# Using | operator
result = set1 | set2
print("Union of sets using | operator:", result)  # Output: {1, 2, 3, 4, 5}
```

* str.isupper()
* Returns True if all cased characters in the string are uppercase and there is at least one cased character. Non-cased characters are ignored.
* str.islower()
* Returns True if all cased characters in the string are lowercase and there is at least one cased character. Non-cased characters are ignored.

```python
# Example with uppercase string
s1 = "HELLO!"
print(s1.isupper())   # Output: True
print(s1.islower())   # Output: False

# Example with lowercase string
s2 = "hello!"
print(s2.isupper())   # Output: False
print(s2.islower())   # Output: True

# Example with mixed case and special characters
s3 = "Hello World!"
print(s3.isupper())   # Output: False
print(s3.islower())   # Output: False

# Example with digits and special characters only
s4 = "1234!@#$"
print(s4.isupper())   # Output: False
print(s4.islower())   # Output: False

# Example with mixed case, digits, and special characters
s5 = "HELLO123!@#"
print(s5.isupper())   # Output: True
print(s5.islower())   # Output: False

s6 = "hello123!@#"
print(s6.isupper())   # Output: False
print(s6.islower())   # Output: True
```

### Copy v/s Deepcopy
* The copy method creates a shallow copy of an object. This means it creates a new object but inserts references to the objects found in the original. In other words, it copies the top-level structure of the object, but not the nested objects themselves.
* The deepcopy function, on the other hand, creates a complete copy of an object and all of its nested objects, recursively. It means it duplicates everything, including any objects inside the original object.
##### When to Use Which:
* Use copy when you want to create a new object with a structure similar to the original but don't need to duplicate the nested objects. It's efficient for simple structures.

* Use deepcopy when you need a completely independent copy of an object and all of its nested objects. This is necessary when working with complex data structures or when you want to modify one copy without affecting the other.

```python
import copy

# Original list with nested lists
original_list = [[1, 2, 3], [4, 5, 6]]

# Shallow copy
shallow_copy_list = copy.copy(original_list)

# Modify the shallow copy
shallow_copy_list[0][0] = 100

print("Original list:", original_list)
print("Shallow copy list:", shallow_copy_list)
```

```python
import copy

# Original list with nested lists
original_list = [[1, 2, 3], [4, 5, 6]]

# Deep copy
deep_copy_list = copy.deepcopy(original_list)

# Modify the deep copy
deep_copy_list[0][0] = 100

print("Original list:", original_list)
print("Deep copy list:", deep_copy_list)
```

```python
#try,except,finally,else
def divide(x, y):
    try:
        result = x / y
    except ZeroDivisionError:
        print("Error: Division by zero!")
    except TypeError:
        print("Error: Unsupported operand type(s) for division!")
    else:
        print(f"Division successful! Result: {result}")
    finally:
        print("Execution completed.")

# Example usage:
divide(10, 2)    

```
