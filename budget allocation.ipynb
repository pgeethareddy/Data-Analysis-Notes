```python
plt.figure(figsize=(20,10))

sns.histplot(data=df,x='height',hue='gender')

plt.axvline(df["height"].mean(), color='coral', label='Mean')

plt.axvline(df["height"].mean() + (1 * df["height"].std()), color='yellow', label='Mean + 1SD')
plt.axvline(df["height"].mean() - (1 * df["height"].std()), color='yellow', label='Mean - 1SD')
plt.axvline(df["height"].mean() + (2 * df["height"].std()), color='green', label='Mean + 2SD')
plt.axvline(df["height"].mean() - (2 * df["height"].std()), color='green', label='Mean - 2SD')
plt.axvline(df["height"].mean() + (3 * df["height"].std()), color='blue', label='Mean + 3SD')
plt.axvline(df["height"].mean() - (3 * df["height"].std()), color='blue', label='Mean - 3SD')

plt.legend()

plt.show()
```
